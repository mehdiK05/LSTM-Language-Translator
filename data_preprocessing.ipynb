```python
from datasets import load_dataset
import pandas as pd
from huggingface_hub import login
login()

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("atlasia/Atlaset")

df = ds["train"].to_pandas()  # Convert the "train" split to pandas DataFrame
df
```

```python
df_filtered = df[df["dataset_source"] != "MAC Corpus"]
df_filtered.tail()
```

```python
unique_sources = df["dataset_source"].unique()
unique_sources, len(unique_sources)
```

#Visualisation of the sources

```python
all_tails = []  # Create an empty list to store DataFrames

for source in unique_sources:
    tail_df = df[df["dataset_source"] == source].tail(1)  # Get the tail for each source
    all_tails.append(tail_df)  # Add the tail DataFrame to the list

final_df = pd.concat(all_tails)  # Combine all tail DataFrames into one
final_df  # Display the final DataFrame
```

#Removing rows for other tasks than machine translation

```python
all_tails = []  # Create an empty list to store DataFrames

for i in range(20):
    if unique_sources[i] != 'bourbouh/moroccan-darija-youtube-subtitles':
      tail_df = df[df["dataset_source"] == unique_sources[i]]  # Get the tail for each source
      all_tails.append(tail_df)  # Add the tail DataFrame to the list

final_df = pd.concat(all_tails)  # Combine all tail DataFrames into one
final_df  # Display the final DataFrame
```
